```python
import matplotlib.pyplot as plt
import numpy as np
stu_name=['samsung', 'pepsi', 'nasa', 'pablo escobar', 'donald trump']
stu_marks=[10, 18, 70, 22, 0]
per=[]
for x in stu_marks:
  res=(x/100)*100
  per.append(res)
print(per)
def line_chart():
  plt.plot(stu_name, stu_marks)
  plt.title('marks of students')
  plt.xlabel('students')
  plt.ylabel('marks')
  plt.show()
line_chart()
def bar_chart():
  plt.bar(stu_name, stu_marks)
  plt.title('marks of students')
  plt.xlabel('students')
  plt.ylabel('marks')
  plt.show()
bar_chart()
```
