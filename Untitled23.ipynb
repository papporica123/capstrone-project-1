```python
import matplotlib.pyplot as plt
blood_sugar_men = [113, 85, 90, 150, 149, 89, 105]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 87]
type=[blood_sugar_men,blood_sugar_women]
color=['b','r']
label=['men','women']
bin=[80,100,125,150]
plt.xlabel('blood_sugar_range')
plt.ylabel('blood_sugar_patients')
plt.hist(type,bins=bin,rwidth=0.95,color=color,label=label,orientation='horizontal')
plt.legend()
plt.show()
```
