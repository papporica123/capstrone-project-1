```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('Titanic Dataset (1).csv')
data.head(15)
```

```python
sns.boxplot(data=data,x='Pclass',y='Age')
plt.show()
```

```python
plt.scatter(x=data['Fare'],y=data['Name'])
plt.show()
```

```python
association_catagorical=pd.crosstab(data['Age'],data['Survived'])
print(association_catagorical)
```

```python
data.columns
```

```python
minimum_age=data['Age'].min()
maximum_age=data['Age'].max()
print(minimum_age)
print(maximum_age)
```

```python
bins=[0,15,30,45,60,75,90]
age_labels=['Children','Youth','Young Adult','Adult','Senior','Old']
data['Age']=pd.cut(data['Age'],bins=bins,labels=age_labels,right=False)
print(data['Age'])
```

```python
plt.distplot(data['Age'])
plt.show()
print(data['Age'].skew())
```
