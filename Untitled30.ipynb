```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('Titanic Dataset.csv')
data.head(11)
```

```python
sns.distplot(data['Age'],kde=False,bins=40)
plt.show()
```

```python
sns.countplot(x='Name',data=data,hue='Parch',palette='mako')
plt.show()
```

```python
sns.barplot(x='Gender',y='Age',data=data,palette='spring')
plt.show()
```

```python
sns.heatmap(data.corr())
plt.show()
```
